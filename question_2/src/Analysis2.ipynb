```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
d_f = pd.read_csv('../PDS_Assignment/StudentsPerformance.csv')
d_f
```

```python
d_f["Percentage"]=(d_f["math_score"]+d_f["reading_score"]+d_f["writing_score"])/3
print(d_f["Percentage"])
```

```python
def Grade(Percentage):
    if (Percentage >= 95):return 'O'
    if (Percentage >= 81):return 'A'
    if (Percentage >= 71):return 'B'
    if (Percentage >= 61):return 'C'
    if (Percentage >= 51):return 'D'
    if (Percentage >= 41):return 'E'
    else: return 'F'
    
d_f["grade"] = d_f.apply(lambda x : Grade(x["Percentage"]),axis=1)
```

```python
d_f
```

```python
sns.barplot(x ='gender', y ='Percentage', data = d_f,
            palette ='plasma', estimator = np.std)
```

```python
d_f['gender'].value_counts()
```

```python
label=['Female', 'Male']

plt.pie(d_f['gender'].value_counts(),labels=label,explode=[0.1,0.1],
        autopct='%1.1f%%',colors=['pink','green'], startangle=90)
```
