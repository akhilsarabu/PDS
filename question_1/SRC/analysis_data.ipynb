```python
import pandas
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
d_f = pandas.read_csv('../PDS_Assignment/clean_frailty.csv', index_col=0)
```

```python
ax = sns.heatmap(d_f.corr(), cmap="YlGnBu", linewidths = 0.30, annot = True)
fig = ax.get_figure()
fig.savefig("../PDS_Assignment/results/heatmap.png")
```

```python
ax=sns.distplot(d_f['age'], kde = False, color ='red', bins = 30)
fig=ax.get_figure()
fig.savefig("../PDS_Assignment/results/Age_Distibution.png")
```

```python
ax=sns.distplot(d_f['grip_strength'], kde = False, color ='blue', bins = 30)
fig=ax.get_figure()
fig.savefig("../PDS_Assignment/Grip_Strength_Distibution.png")
```

```python
ax = d_f.plot.scatter(x='age',y='grip_strength')
fig=ax.get_figure()
fig.savefig("../PDS_Assignment/results/Grip_Strength_Age_Relation.png")
```

```python
ax = d_f.plot.scatter(x='weight',y='grip_strength')
fig=ax.get_figure()
fig.savefig("../PDS_Assignment/results/GripStrength_Weight_Relation.png")
```
